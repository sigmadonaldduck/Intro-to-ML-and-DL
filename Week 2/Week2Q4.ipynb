```python
import pandas as pd

data = pd.read_csv('/kaggle/input/glass/glass.csv')

X = data.drop(['Type', 'Id'], axis=1)  # Exclude 'Type' and 'Id' columns
Y = data['Type']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

import numpy as np

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2, axis=1))

class KNN:
    def __init__(self, k):
        self.k = k
    
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
    
    def predict(self, X):
        y_pred = [self._predict(x) for x in X]
        return np.array(y_pred)
    
    def _predict(self, x):
        
        distances = euclidean_distance(x, self.X_train)
        
        k_indices = np.argsort(distances)[:self.k]
        
        k_nearest_labels = self.y_train.iloc[k_indices]
    
        most_common = np.bincount(k_nearest_labels).argmax()
        
        return most_common

X_train = np.array(X_train)
y_train = np.array(y_train)

knn = KNN(k=3)
knn.fit(X_train, y_train)

from sklearn.metrics import accuracy_score

X_test = np.array(X_test)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

from sklearn.neighbors import KNeighborsClassifier

sklearn_knn = KNeighborsClassifier(n_neighbors=3)
sklearn_knn.fit(X_train, y_train)

sklearn_y_pred = sklearn_knn.predict(X_test)

sklearn_accuracy = accuracy_score(y_test, sklearn_y_pred)
print("Sklearn KNN Accuracy:", sklearn_accuracy)

from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

dt_classifier = DecisionTreeClassifier(max_depth=5)

dt_classifier.fit(X_train, y_train)

fig, ax = plt.subplots(figsize=(12, 12))
tree.plot_tree(dt_classifier, feature_names=X.columns, class_names=dt_classifier.classes_, filled=True, ax=ax)
plt.show()
```
